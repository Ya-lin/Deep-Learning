### Diffusion schedules

```python
class diffusion_schedule:

    def linear(self, diffusion_times):
        min_rate = 0.0001
        max_rate = 0.02
        rate_diff = max_rate - min_rate
        df_tensor = torch.tensor(diffusion_times, dtype=torch.float32)
        betas = min_rate + df_tensor*rate_diff
        alphas = 1-betas
        alpha_bars = torch.cumprod(alphas, dim=0)
        signal_rates = alpha_bars
        noise_rates = 1-alpha_bars
        return signal_rates, noise_rates

    def cosine(self, diffusion_times):
        df_tensor = torch.tensor(diffusion_times, dtype=torch.float32)
        signal_rates = torch.cos(df_tensor*np.pi/2)
        noise_rates = torch.sin(df_tensor*np.pi/2)
        return signal_rates, noise_rates

    def offset_cosine(self, diffusion_times):
        min_signal_rate = torch.tensor(0.02).float()
        max_signal_rate = torch.tensor(0.95).float()
        start_angle = torch.acos(max_signal_rate)
        end_angle = torch.acos(min_signal_rate)
        angle_diff = end_angle - start_angle
        df_tensor = torch.tensor(diffusion_times, dtype=torch.float32)
        df_angles = start_angle + df_tensor*angle_diff
        signal_rates = torch.cos(df_angles)
        noise_rates = torch.sin(df_angles)
        return signal_rates, noise_rates  
```

```python
T = 1000
diffusion_times = [x/T for x in range(T)]
ds = diffusion_schedule()
linear_signal_rates, linear_noise_rates  = ds.linear(diffusion_times)
cosine_signal_rates, cosine_noise_rates  = ds.cosine(diffusion_times)
offset_cosine_signal_rates, offset_cosine_noise_rates  = ds.offset_cosine(diffusion_times)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1,2,1)
plt.plot(diffusion_times, linear_signal_rates**2, linewidth=1.5, label="linear")
plt.plot(diffusion_times, cosine_signal_rates**2, linewidth=1.5, label="cosine")
plt.plot(diffusion_times, offset_cosine_signal_rates**2, linewidth=1.5, label="offset_cosine")
plt.xlabel("t/T", fontsize=12)
plt.ylabel(r"$\bar{\alpha_t}$ (signal)", fontsize=12)
plt.title('Signal')
plt.legend()
plt.subplot(1,2,2)
plt.plot(diffusion_times, linear_noise_rates**2, linewidth=1.5, label="linear")
plt.plot(diffusion_times, cosine_noise_rates**2, linewidth=1.5, label="cosine")
plt.plot(diffusion_times, offset_cosine_noise_rates**2, linewidth=1.5, label="offset_cosine")
plt.xlabel("t/T", fontsize=12)
plt.ylabel(r"$1-\bar{\alpha_t}$ (noise)", fontsize=12)
plt.title('Noise')
plt.legend()
plt.show()
```
